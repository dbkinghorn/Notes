```python
import pandas as pd  # data handeling
import numpy as np   # numeriacal computing
import matplotlib.pyplot as plt  # plotting core
import seaborn as sns  # higher level plotting tools
%matplotlib inline
sns.set()
```

```python
df = pd.DataFrame({'CPU':[
    'Intel Xeon-W 2175 14-Core \n 2.5GHz (4.3GHz Turbo)',
    'AMD TR 2990WX 32-Core \n 3.0GHz (4.2GHz Turbo)'
     ], 
    'GFLOP/s':[838, 597]})
```

```python
df
```

```python
plt.figure(figsize=(9,2))

ax = sns.barplot(y="CPU", x="GFLOP/s", data=df, palette="Greens")
#ax.set_xlim(100,320)
ax.set_title('HPL Linpack Benchmark Threadripper 2990WX vs Xeon-W 2175 \n (higher is better)', fontsize=18)

y = df['GFLOP/s']
for i, v in enumerate(y):
    ax.text(v , i + .125, str(v), color='black', fontweight='bold')
```
