```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
import seaborn as sns
sns.set() # not using seaborn but this makes the plots look better
%matplotlib inline
```

```python
w=np.array([
180,
190,
200,
210,
220,
230,
240,
250,
260,
270,
280,
290,
300,
310,
320,
330,
340,
350,])
```

```python
fp32x1=np.array([
236.8,
264.0,
305.2,
333.9,
372.0,
402.5,
434.8,
460.2,
479.7,
494.3,
508.6,
519.0,
525.4,
531.1,
535.0,
540.5,
542.5,
544.3,])
```

```python
def fcn(x,a,b,c):
    return a  + b*x + c*np.log(1+x)
    #return a + b*np.log2(c*x+d)
popt, pcov = curve_fit(fcn, w, fp32x1)
popt
```

```python
plt.rcParams["figure.figsize"] = [12,7]
fig, ax = plt.subplots()
ax.plot( w, fp32x1, "o", color='g')  # plot the test data
xt = np.linspace(160,350,100)
ax.plot(xt, fcn(xt,*popt) )  # plot the model function
#  ,  label='a/((1-b)+(b/n)) with a=%2.6f b=%2.6f n=%2.6f' % tuple(popt)
max95 = fp32x1[-1]*0.95
xp = np.linspace(max95,max95,100)
ax.plot(xt,xp, color='y', label='95% of peak')

plt.xlabel("Watts")
plt.ylabel("images/sec")
plt.title("1 x RTX3090 TensorFlow ResNet50 fp32 -- Watts vs Performance", fontsize=18)
ax.legend()
```
