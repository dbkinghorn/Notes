```python
def fcn(x,a,b,c):
    return a  + b*x + c*np.log(1+x)
    #return a + b*np.sqrt(x) * np.log(x+1)
popt32x1, pcov = curve_fit(fcn, w, fp32x1)
popt32x1
```

```python
plt.rcParams["figure.figsize"] = [12,7]
fig, ax = plt.subplots()
ax.plot( w, fp32x1, "o", color='g')  # plot the test data
xt = np.linspace(160,350,100)
ax.plot(xt, fcn(xt,*popt32x1) )  # plot the model function
#  ,  label='a/((1-b)+(b/n)) with a=%2.6f b=%2.6f n=%2.6f' % tuple(popt)
max95 = fp32x1[-1]*0.95
xp = np.linspace(max95,max95,100)
ax.plot(xt,xp, color='y', label='95% of peak')

plt.xlabel("Watts")
plt.ylabel("images/sec")
plt.title("1 x RTX3090 TensorFlow ResNet50 fp32 -- Watts vs Performance", fontsize=18)
ax.legend()
```

```python
fp32 = np.array([544.3,969.2,1434.3,1897.9])
pl32 = np.array([508.6,922.5,1360.3,1803.7])
#fp32 = fp32[0]/fp32
numgpus = np.array([1,2,3,4])
fp32
```

```python
def amdhal1(n,P):
    return fp32[0]/((1-P)+(P/n))

poptfp, pcov = curve_fit(amdhal1, numgpus, fp32)

def amdhal2(n,P):
    return pl32[0]/((1-P)+(P/n))
poptpl, pcov = curve_fit(amdhal2, numgpus, pl32)
```

```python
plt.rcParams["figure.figsize"] = [12,7]
#plt.figure(figsize=(16,9))
fig, ax = plt.subplots()
ax.plot( numgpus, fp32, "o", color='g', label='ResNet50 img/sec')  # plot the test data
ax.plot( numgpus, pl32, "o", color='r', label='At 280W per GPU power limit') 
#ax.plot( numgpus, stmv32, "x", color='r', label='NAMD STMV: "Wall Time 3970x')
#ax.plot(4, 1803.7, "X", color='r', label='At 280W per GPU power limit')

xt = np.linspace(0.5,8,20)
ax.plot(xt, amdhal1(xt,poptfp)  ,  label='Amdhals Eqn with P = %.4f ' %(poptfp[0]))  # plot the model function
#ax.plot(xt[:11], amdhal32(xt[:11],popt32)  ,  label='Amdhals Eqn with P = %.4f ' %(popt32[0]))
ax.plot(xt, amdhal2(xt,poptpl)  ,  label='Amdhals Eqn with P = %.4f ' %(poptpl[0]))

ax.plot(xt,fp32[0]*xt, color='k', label='Linear Scaling')

plt.xlabel("Number of GPUs")
plt.ylabel("Images/Second")
plt.title("Amdhal's Law, RXT3090 1-4 GPU Scaling \n TensorFlow 1.5 ResNet50 fp32", fontsize=18)
ax.legend()
```
